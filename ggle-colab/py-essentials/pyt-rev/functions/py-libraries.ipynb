```python
import pandas as pd

# create some random data

var1 = np.random.randn(100)*5 + 20
var2 =  np.random.randn(100)>0

labels = ['Temp. (C)', 'Ice cream  eaten']

# create a dictionary

D = {labels[0]: var1, labels[1]: var2 }


# import the dictionary into a pandas dataframe


df = pd.DataFrame(data=D)
df
```

```python
# Exercises: Create a pandas dataframe with three variables:
# 1) integers from 0 to 10
# 2) their square
# 3) their log
```

```python
nums = np.array(range(11))
nums

sqr = np.square(nums)
sqr

logs = np.log(nums)
logs

labels = [ 'Integer', 'Square root', 'logs']

D = { labels[0]: nums, labels[1]: sqr, labels[2]: logs}

df = pd.DataFrame(data=D)
df
```
