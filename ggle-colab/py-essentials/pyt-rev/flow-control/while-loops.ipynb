```python
# Exercises
# make a Poisson-like counter
# parameter lamba. positive number(10)
# start at 1, multiply by random number uniform([0, 1])
# target number is smaller than exp(-lambda)
# count the number of iterations until target<exp(-lamba)
```

```python
import numpy as np

def poissonCounter(lam):
  # initialize some variables
  counter, currVal = 0, 1
  target = np.exp(-lam)

  # run algorithms
  while currVal > target:
    counter += 1
    currVal *= np.random.rand()

  # return the result (counter)
  return counter


```

```python
poissonCounter(10)
```

```python
p = np.zeros(100)

for i in range(100):
  p[i] = poissonCounter(10)

p
```
