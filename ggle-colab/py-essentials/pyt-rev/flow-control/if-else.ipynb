```python
# Exercises
# Create a dot product function
# [1, 2], [3, 0] -> (1*3) + (3*0) -> 3 + 0 -> 3
# functions checks that both inputs are numpy arrays (hint: isinstance())
# function check for same length of two input vectors
# returns dot product if pass, useful error message otherwise (hint: raise)
```

```python
def dotproduct(a, b):
  # check that both vectors are numpy array
  if not (isinstance(a, np.ndarray) and isinstance(b, np.ndarray)):
    raise Exception('Inputs must be numpy arrays')

  # check that both vectors have the same length
  if not len(a) == len(b):
    raise Exception('Must be the same length')

  # return the dot product
  return sum(a*b)
```

```python
v = np.array([3,4,2])
w = np.array([3,4,1])

print(dotproduct(v, w))
print(np.dot(v, w))
```
