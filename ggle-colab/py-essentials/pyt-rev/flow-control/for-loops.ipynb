```python
# Exercise

# create a function that reports a Fibonacci serieus for n+3 numbers

# 0,1,1,2,3,5,8, ...

def fibseq(n):
  v = [0,1,1]
  for i in range(n):
    v.append(sum(v[-2:]))

  return v
```
