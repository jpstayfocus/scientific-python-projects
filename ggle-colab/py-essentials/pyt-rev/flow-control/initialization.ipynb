```python
import numpy as np

n = 20

r = np.zeros(n)

for i in range(n):
  r[i] = i**2

print(r)
```

```python
# del r

r = []

for i in range(n):
  r.append(i**2)

print(r)
```

```python
# del r

r = np.array([])

for i in range(n):
  r = np.append(r, i**2)

print(r)
```
