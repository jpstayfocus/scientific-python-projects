# Mean and median

```
# create a dataset
N = 11
data = np.random.randn(N)
```

```
# compute the mean and compare to numpy's mean

mymean = sum(data)/len(data)
npmean = np.mean(data)

print(mymean)
print(npmean)
```

```
# compute the median
# print(data)
data.sort()
# print(data)

# algorithm to compute the median
if N%2==1:
  mymedian = data[N//2]
else:
  mymedian = (data[N//2-1]+data[N//2])/2

# numpy's median
npmedian = np.median(data)

print(mymedian)
print(npmedian)
```

# Frequencies table

```
# an aside on 'enumerate'
l = [5,4,5,'asdf']
for i,j in enumerate(l):
  print(i,j)
```

```
# new dataset of integers
data = list( np.random.randint(1,high=6,size=20) )
data
```

```
# compute the unique values and their counts

uniquevals = list( set(data) )
valcounts = [0]*len(uniquevals)

for d in data:
  idx = [i for i,u in enumerate(uniquevals) if d==u]
  valcounts[idx[0]] += 1

print(uniquevals)
print(valcounts)
```

```
# now make the table (as dictionary)
table = {}

for i in range(len(uniquevals)):
  table[uniquevals[i]] = valcounts[i]

table
```

# Mode

```
# maximum count
maxcount = 0

# loop through table items and find largest count
for i in table.items():
  if i[1] > maxcount:
    maxcount = i[1]
    mode = i[0] # mode is the value with the largest count

print('Mode = %s (appears %s times)'%(mode,maxcount))
```

```
# alternative
maxcount = max(list(table.values()))
mode = [k for k,v in table.items() if v==maxcount]
mode
```

# Standard deviation

```
# a new random dataset

N = 51
data = list( np.random.poisson(2,N) )

meanval = float( sum(data)/len(data) ) # remove numpy type
a = sum([(e-meanval)**2 for e in data])
std = (a/(N-1))**(1/2)

print(std)
print(np.std(data))
```
